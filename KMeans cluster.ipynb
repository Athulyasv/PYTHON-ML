```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
data = pd.read_csv('/content/drive/My Drive/DATASETS/sales data-set.csv')
data
datax=data.iloc[:,0:2].values.reshape(-1,1)
datax
datay=data.iloc[:,3].values
datay
kms =KMeans(n_clusters=3,max_iter=1000)
kms.fit(datax,datay)
p = kms.predict(datax)
plt.scatter(datax,p, color = "red")
```
